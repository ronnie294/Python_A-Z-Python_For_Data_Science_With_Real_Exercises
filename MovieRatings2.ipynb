```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

```python
Movies.describe()
```

```python
List1=['action','comedy','adventure','animation','drama']
List2=['Buena Vista Studios','Sony','Universal','WB','Paramount Pictures','Fox']
```

```python
Filtered=Movies[(Movies.Genre=='action')|(Movies.Genre=='comedy')|(Movies.Genre=='adventure')|(Movies.Genre=='animation')|(Movies.Genre=='drama')]
Filtered2=Filtered[(Filtered.Studio=='Buena Vista Studios')|(Filtered.Studio=='Sony')|(Filtered.Studio=='Universal')|(Filtered.Studio=='WB')|(Filtered.Studio=='Paramount Pictures')|(Filtered.Studio=='Fox')]
```

```python
plt.rcParams['figure.figsize']=10,10
sns.set(palette='muted',color_codes=True)
K1=sns.boxplot(data=Filtered2,x='Genre',y='Gross % US',width=0.8,color='lightgray')
K1=sns.stripplot(data=Filtered2,x='Genre',y='Gross % US',jitter=True,hue='Studio')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Genre',fontsize=20)
plt.ylabel('Gross % US',fontsize=20)
plt.title('Domestic Gross % by Genre',fontsize=20)


```
